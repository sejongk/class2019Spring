```python
# description
# https://www.kaggle.com/c/boston-housing

# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation # full connection
#activation에 따라 성능 달라짐
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

# train 과 test set으로 나누기

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
# x는 
x_train.shape, x_test.shape
# 13개 벡터에 데이터가 404개 있다.
```

# Regression( continuous - continuous)

#### 13개 input 노드 -> 1개의 노드를 가진 hidden layer (relu적용)
#### Param # = 14 => input 13개 + 상수항 1개
##### x1... x13, 1(b)

#### model.add: 모델에 layer 추가하기, 노드 수, activation을 option으로 줄 수 있다.

```python
model = Sequential()
model.add(Dense(1, input_shape=[13], activation = 'relu'))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

# 모델 학습

#### epochs : train 데이터을 돌리는 횟수

```python
model.fit(x_train, y_train, epochs=1000)
#loss 의미?
```

# 모델 평가

### test set을 이용해서 모델의 loss를 출력한다.

```python
model.evaluate(x_test, y_test)
```

# hidden layer추가를 통한 성능향상

#### 노드 수: 14 (13+1) -> 11 (10+1) ->21 (20+1) -> 2 (1+1)
#### Param #: 14 * 10 = 140, 11 * 20 = 220

```python
model = Sequential()
model.add(Dense(10, input_shape=[13],activation = 'relu'))
model.add(Dense(20)) # 20이여도 상수인 노드가 하나 더 있어서 이 layer에는 21개의 node가 있다.
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
model.fit(x_train, y_train, epochs=1000)
#linear하기 때문에 hidden layer가 있는것과 없는 것끼리 차이가 없다.
#parameter 는 y=ax+b에서 a와 b가 parameter 
'''
이 코드에서 hidden layer 2개
화살표의 개수 = parameter # 
10 * (13+1) = 140
hidden layer의 b에는 그 이전 layer로 부터 오는 화살표는 없지만 다음 layer로 나가는 화살표는 있다.
(10+1) * 20 = 220
20+1 = 21
'''
```

### 두 개의 hidden layer를 relu로 설정해 성능향상

#### output layer에서는 regression 아닐 때 non- linear로 
#### softmax=> multiple category class
#### sigmoid=> binary class

```python
model = Sequential()
model.add(Dense(10, input_shape=[13],activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
model.fit(x_train, y_train, epochs=1000)
```
