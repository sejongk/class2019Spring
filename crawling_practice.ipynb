## 실습: 네이버 뉴스, 분야 별 뉴스 랭킹 가져오기
순위,기사제목, 신문사, 조회수 출력

```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://news.naver.com/main/ranking/popularDay.nhn')
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

```python
kind = ["정치","경제","사회","생활/문화","세계","IT/과학"]
print("<"+"랭킹: 많이 본 뉴스"+">")
for i in range(0,6):
    category = bsObject.find_all("ol")[i]
    print("["+kind[i]+"]")
    
    for n in range(1,6):
        li = category.find("li", {"class":"num"+str(n)})
        dt = li.select("dt")[0] #find_all = select
        title = dt.select('a')[0].get('title')
        publisher = li.get_text().split()[-3]
        view = li.get_text().split()[-1]
        print(n ,title, publisher, view)

    
```

```python
# Create a file to write to, add headers row
f = csv.writer(open('newsRanking.csv', 'w'))
kind = ["정치","경제","사회","생활/문화","세계","IT/과학"]
f.writerow(["<"+"랭킹: 많이 본 뉴스"+">"])

for i in range(0,6):
    f.writerow(["["+kind[i]+"]"])
    f.writerow(['Title', 'Newspaper', "Views"])
    category = bsObject.find_all("ol")[i]
    for n in range(1,6):
        li = category.find("li", {"class":"num"+str(n)})
        dt = li.select("dt")[0] #find_all = select
        title = dt.select('a')[0].get('title')
        publisher = li.get_text().split()[-3]
        view = li.get_text().split()[-1]
        f.writerow([n ,title, publisher, view])
```
