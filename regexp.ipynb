```python
import nltk
import re #regular expression. 
```

```python
''' 
 Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures
.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]      Matches anything but a set of characters 
=>caret이 []안에 들어가있으면 제한시키는 역할을 한다.
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	        Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	    Parentheses that indicate the scope of the operators
'''
```

### re.search(조건, 대상) -> 대상에서 조건이 만족하는지 check

```python
# re.search(조건, 대상)
print(re.search('abc','abcde')) 
#regular expression은 list가 아닌 string에 적용이 되고 list전체에 적용시키려면 for loop를 돌면서 적용해야한다.
```

```python
'''
^abc	    처음부분이 abc와 일치
abc$	    뒷부분이 abc와 일치
^abc$	    처음과 뒷부분이 abc와 일치, abc로 시작, abc로 종료
'''
```

```python
re.search('^ab$','abcde') #조건을 만족하지 않으므로 return 값 없음
```

#### ed$ -> ed로 끝나는가

```python
result = [w for w in wordlist if re.search('ed$', w)][:10] 
print(result)
```

#### .  -> wildcard 아무 하나의 문자면 ok

```python
result = [w for w in wordlist if re.search('^..j..t..$', w)][:10] # . -> wildcard 아무 문자 , ^ -> caret
print(result[:10])
```

#### [] ->[]에 들어간 아무 알파벳 중 하나만 돼야됨

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)] 
print(result[:10])
```

#### + -> 앞에 나온 조건이 1회 이상 적용가능

```python
result = [w for w in wordlist if re.search('^[ah]+$', w)][:10] # +는 1회 이상 match
print(result[:10]) # [ah] 조건이 여러번 반복되어도됨.
```

#### \. -> 특수문자 점(.)을 의미 , [0-9] -> 0부터 9까지 아무 숫자면 가능

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)] # \. -> 점(.)을 의미
print(re.search('[0-9]+\.[0-9]+$','a0.001'))
print(re.search('^[0-9]+\.[0-9]+$','a0.001'))
print(wordlist[:100])
```

#### [A-Z] -> 대문자 A부터 Z까지 아무 알파벳이나 가능, [A-Za-z] = [a-zA-Z] -> 알파벳 대소문자

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)] # \$ -> 특수문자 $ 표현
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[a-z]+\$$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[A-Za-z]+\$$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[^C]+\$$', w)]
print(wordlist[:10])
```

#### {k} -> 앞 조건이 k번 반복 적용되면 ok

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)] # 0-9조건이 4번 반복 총 4글자
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[0-9]{3}$', w)] # 0-9조건이 3번 반복 3자리 숫자
print(wordlist[:10])
```

#### - -> 하이픈 표시, {m, n} -> 앞 조건이 m번부터 n번까지 반복되어서 적용되면 ok

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)] #{m,n} m번 부터 n번까지 okay , -는 그냥 hyphen
print(wordlist[:10])
```

#### {m, } -> m번 이상 , {,m} -> m번 이하 반복되어서 적용되면 ok

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

#### | -> or , 조건들 중 하나만 만족하면 ok

```python
# ed로 끝나거나 ing로 끝나는 것들

wordlist = [w for w in wsj if re.search('(ed|ing)$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('(ed$|ing$)', w)]
print(wordlist[:10])
```

## re.findall()

#### string에서 패턴을 만족하는 모든 문자열을 추출해 list로 만든다.

```python
word = 'supercalifragilisticexpialidocious'
result = re.findall(r'[aeiou]', word)
result = re.findall(r'[li]{1,3}', word)

print(result)
```

#### (..) -> 조건을 만족하는 length가 '.' 개수인 문자열을 반환

```python
result = re.findall('[aeiou](...)[aeiou]', word) 
print(result)
result = re.findall('[aeiou](..)[aeiou]', word) 
print(result)
result = re.findall('[aeiou]..[aeiou]', word) 
print(result)
```

#### [^abc] -> abc를 제외한 나머지 문자들

```python
result = re.findall('[^aeiou].+[^aeiou]', word)
print(result)
result = re.findall('[^aeiou].+?[^aeiou]', word) # +? 란 무엇일까?
print(result)
result = re.findall('[^aeiou].?[^aeiou]', word)
print(result)
```

#### nltk.FreqDist() -> list의 원소들의 freq를 반환한다.

generator vs list

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
print(type(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word)))
print(type([vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word)]))

print(fdist.most_common(11))
print([str for (str, freq) in fdist.most_common(11)])
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))

li =['a','b','v']
st = 'str'
st.join(li)
```
