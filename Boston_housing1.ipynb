```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
```

#### Boston_housing Data를 불러와 train, test set으로 나누기

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
len(x_train), len(x_test)
```

### Modlel 설정(regression)- Sequential()

#### Dense(출력개수, input_shape=[입력개수)

#### model.compile(optimizer, loss) - optimizer와 loss 계산 방법 설정
#### model.summary()- model 정보를 보여준다.

```python
model = Sequential()
model.add(Dense(1, input_shape=[13])) # input shape = 13, 출력이 1개
model.compile(optimizer = 'rmsprop',loss='mse')
model.summary()
```

### Model 학습 - model.fit

#### epochs - 학습 횟수

```python
model.fit(x_train, y_train, epochs =10000)
```

### 모델 평가 - model.evaluate() - 출력값: loss

```python
model.evaluate(x_test, y_test)
```
